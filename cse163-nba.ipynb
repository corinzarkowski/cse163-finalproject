```python
player_data_formatted_years = player_data_formatted_cbb.copy()

for year in range(1950, 2022):
    print(year)
    URL = 'https://www.basketball-reference.com/leagues/NBA_' + str(year) + '_totals.html'
    data = requests.get(URL, stream=True)
    for line in data.iter_lines():
        match = re.findall('data-stat="player" csk="[^"]*" ><a href="/players/[^>]*>([^<]*)</a></td>', str(line))
        match2 = re.findall('class="italic_text partial_table"', str(line))
        if(match and not match2 and 
           match[0] in player_data_formatted_years.keys() and 
           'Games' in player_data_formatted_years[match[0]].keys()):

            player_data_formatted_years[match[0]][year] = {}

            matchPoints = re.findall('data-stat="pts" >([\d]*)</td>', str(line))
            if matchPoints:
                player_data_formatted_years[match[0]][year]['points'] = matchPoints[0]

            matchRebounds = re.findall('data-stat="trb" >([\d]*)</td>', str(line))
            if matchRebounds:
                player_data_formatted_years[match[0]][year]['rebounds'] = matchRebounds[0]

            matchSteals = re.findall('data-stat="stl" >([\d]*)</td>', str(line))
            if matchSteals:
                player_data_formatted_years[match[0]][year]['steals'] = matchSteals[0]

            matchAssists = re.findall('data-stat="ast" >([\d]*)</td>', str(line))
            if matchAssists:
                player_data_formatted_years[match[0]][year]['assists'] = matchAssists[0]

            matchBlocks = re.findall('data-stat="blk" >([\d]*)</td>', str(line))
            if matchBlocks:
                player_data_formatted_years[match[0]][year]['blocks'] = matchBlocks[0]
```

```python
with open("full_data_career_stats.json", "w") as outfile:
    json.dump(player_data_formatted_years, outfile)
```

```python
player_data_final = player_data_formatted_years.copy()

for player_name in list(player_data_formatted_years.keys()):
    best_year = -1
    cur_year = 1
    stat_total = 0
    cur_stat_total = 0
    
    for year_or_other in player_data_formatted_years[player_name]:
        if type(year_or_other) == int:
            stats = []
            for stat in player_data_formatted_years[player_name][year_or_other]:
                stats.append(player_data_formatted_years[player_name][year_or_other][stat])
            
            if '' not in stats:
                cur_stat_total = int(player_data_formatted_years[player_name][year_or_other]['points']) * \
                                 int(player_data_formatted_years[player_name][year_or_other]['rebounds']) * \
                                 int(player_data_formatted_years[player_name][year_or_other]['steals']) * \
                                 int(player_data_formatted_years[player_name][year_or_other]['assists']) * \
                                 int(player_data_formatted_years[player_name][year_or_other]['blocks'])
                if cur_stat_total > stat_total:
                    stat_total = cur_stat_total
                    best_year = cur_year
                cur_year += 1
    
    player_data_final[player_name]['best_year'] = best_year
```

```python
player_data_by_year = {}

for year in range(1950, 2022):
    print(year)
    URL = 'https://www.basketball-reference.com/leagues/NBA_' + str(year) + '_totals.html'
    data = requests.get(URL, stream=True)
    for line in data.iter_lines():
        match = re.findall('data-stat="player" csk="[^"]*" ><a href="/players/[^>]*>([^<]*)</a></td>', str(line))
        match2 = re.findall('class="italic_text partial_table"', str(line))
        
        if match and not match2 and match[0] in players.keys():
            if match[0] not in player_data_by_year.keys():
                player_data_by_year[match[0]] = {}
            
            player_data_by_year[match[0]][year] = {}

            matchPoints = re.findall('data-stat="pts" >([\d]*)</td>', str(line))
            if matchPoints:
                player_data_by_year[match[0]][year]['points'] = matchPoints[0]

            matchRebounds = re.findall('data-stat="trb" >([\d]*)</td>', str(line))
            if matchRebounds:
                player_data_by_year[match[0]][year]['rebounds'] = matchRebounds[0]

            matchSteals = re.findall('data-stat="stl" >([\d]*)</td>', str(line))
            if matchSteals:
                player_data_by_year[match[0]][year]['steals'] = matchSteals[0]

            matchAssists = re.findall('data-stat="ast" >([\d]*)</td>', str(line))
            if matchAssists:
                player_data_by_year[match[0]][year]['assists'] = matchAssists[0]

            matchBlocks = re.findall('data-stat="blk" >([\d]*)</td>', str(line))
            if matchBlocks:
                player_data_by_year[match[0]][year]['blocks'] = matchBlocks[0]
                
print(len(player_data_by_year.keys()))
```

```python
player_data_noyear = {}

for player_name in player_data_by_year.keys():
    best_year = -1
    cur_year = 1
    stat_total = 0
    cur_stat_total = 0
    
    for year in player_data_by_year[player_name]:
        stats = [player_data_by_year[player_name][year][stat] for stat in player_data_by_year[player_name][year]]

        if '' not in stats:
            cur_stat_total = int(player_data_by_year[player_name][year]['points']) * \
                             int(player_data_by_year[player_name][year]['rebounds']) * \
                             int(player_data_by_year[player_name][year]['steals']) * \
                             int(player_data_by_year[player_name][year]['assists']) * \
                             int(player_data_by_year[player_name][year]['blocks'])
            if cur_stat_total > stat_total:
                stat_total = cur_stat_total
                best_year = cur_year
            cur_year += 1
    
    if best_year != -1:
        player_data_noyear[player_name] = {}
        player_data_noyear[player_name]['best_year'] = best_year
        player_data_noyear[player_name]['career_length'] = cur_year
```
